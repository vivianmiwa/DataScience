```python
avaliacoes = pd.read_csv("notasMovieLens.csv")
avaliacoes.head()
```

```python
avaliacoes.shape # mostra o numero de linhas e colunas
```

```python
avaliacoes.columns = ["usuarioId", "filmeId", "nota", "momento"]
avaliacoes.head()
```

```python
avaliacoes_do_filme_1 = avaliacoes.query("filmeId==1") # mostra apenas as avaliacoes do filme com id 1
avaliacoes_do_filme_1.head()
```

```python
avaliacoes_do_filme_1.describe() # estatisticas detalhadas
```

```python
avaliacoes_do_filme_1.mean() # calcula as médias
```

```python
# calculando a média de apenas da coluna nota
avaliacoes["nota"]
avaliacoes["nota"].mean()
```

```python
# calculando a média das notas do filme 1
avaliacoes_do_filme_1["nota"].mean()
```

```python
# calculando a média das notas de todos os filmes
notas_medias_por_filme = avaliacoes.groupby("filmeId")["nota"].mean() # agrupado por filmes
notas_medias_por_filme.head()
```

```python
filmes
```

```python
notas_medias_por_filme
```

```python
# juntando as notas médias com as informações dos filmes
filmes_com_media = filmes.join(notas_medias_por_filme, on="filmeId")
filmes_com_media.head()
```

```python
# ordenando de forma decrescente
filmes_com_media.sort_values("nota", ascending=False).head(15)
```

```python
# renomeando a coluna nota para media
filmes_com_media = filmes_com_media.rename(columns={"nota": "media"})
filmes_com_media.head()
```

```python
# procurando filmes sem nota (NaN)
filmes_com_media.sort_values("media")
```

```python
# arredondando media para 2 casas decimais
filmes_com_media = filmes_com_media.round({"media": 2})
filmes_com_media
```
